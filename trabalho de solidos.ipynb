Abaixo temos o programa que calcula e plota o fluxo de cisalhamento horizontal:

$$q_h(z) = \cfrac{V~ d~ t}{I_z}(d - t)$$

```python
% matplotlib inline
import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt
z=0
V = 106
Iz = 0.0000018
d = 0.08
t = 0.01

g = lambda z: ((V*z*t)/Iz)*(d-t)


# cria pontos (x, y) da reta 
Xr = np.arange(-3, 3, 0.1)
Yr = []
for x in Xr:
    Yr.append(g(z))    

# Plota os pontos e a reta
plt.plot(Xr,g(Xr),'-') 


plt.xlabel('distância(cm)',fontsize=16)
plt.ylabel('Fluxo de cisalhamento (N/m)',fontsize=16)
plt.plot(2.5, g(2.5), "ro")
print("fluxo de cisalhamento no ponto 2.5 cm:", g(2.5))
plt.show()
```
